```python
#import modul

import matplotlib.pyplot as plt
import numpy as np
```

```python
#membuat ploting

y = [1, 2, 3, 4]
x = [1, 3, 4, 2]

fig, ax = plt.subplots() #fungsi untuk membuat axes dan figure (fig = untuk menampung object figure, ax = untuk menampung objext axes)
ax.plot(x, y)
```

```python
#cara simple membuat ploting menggunakan fungsi "plot()"
y = [1, 2, 3, 4]
x = [1, 3, 4, 2]
plt.plot(x, y)
```

```python
#figure, axes, axis 
#figure = tempat atau kontainer dasar untuk melakukan fisualisasi dasar
#axes = area gambar di dalam figure, dalam satu figure bisa memilki beberapa axes
#axis = sumbu x dan y
```

```python
#dua cara untuk menggunakan matplotlib
#OO Style (Object Oriented Style)
#pyplot style
```

```python
#OO Style

#menyiapkan data set menggunakan =
x = np.linspace(0, 2, 100)
#np.linspace = fungsi yang menghasilkan deret bilangan
#0 = garis start
#2 = garis finish
#100 = jumlah data point
x
```

```python
#ploting menggunakan ax.plot()
flg, ax = plt.subplots()

ax.plot(x, x, label='linear') #nilai x dan y sama 
ax.plot(x, x**2, label='quadratic') #nilai y dihasil kan dari x^2
ax.plot(x, x**3, label='cubic') #nilai y dihasilkan dari x^3

ax.set_xlabel('x label')
ax.set_ylabel('y label')

ax.set_title('simple plot')
ax.legend()
```
